```python
import pandas as pd

df = pd.read_csv ('data_for_notebooks/information_from_links.csv')
df.head(10)
```

```python
del df['Unnamed: 0']
```

```python
df.isnull().sum()
```

```python

df = df.apply(lambda x: x.str.replace('\t',' ')).apply(lambda x: x.str.replace('\n',' '))
```

```python
df = df.apply(lambda x: x.str.replace('\r',' '))

# removing any extra space created because of the above steps
df = df.apply(lambda x: x.str.strip())
df = df.apply(lambda x: x.str.replace('\s{2}',' '))
```

```python
df.head(20)
```

```python
df.to_csv("data_for_notebooks/CLEAN_information_from_links.csv")
```
